```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
from random import uniform
```

```python
class PointHandler:
    i = 1j
    RANGE = (0, 2*np.pi)

    @classmethod
    def get_point(cls):
        point = np.exp(cls.i * uniform(*cls.RANGE))
        x, y = point.real, point.imag
        return x, y

def norm(x, y):
    result = np.linalg.norm(np.array(*x) - np.array(*y))
    return result

def color(distance):
    return "r" if distance < np.sqrt(3)/2 else "b"
```

```python
# TRIANGLE_POINTS = ((0, 1), (np.sqrt(3)/2, -1/2), (-np.sqrt(3)/2, -1/2))
TRIANGLE_POINTS = {"x": (0, np.sqrt(3)/2, -np.sqrt(3)/2, 0), "y": (1, -1/2, -1/2, 1)}
```

```python
a = PointHandler.get_point()
b = PointHandler.get_point()
norm(a, b)
```

```python
figure, axes = plt.subplots(figsize=(8,8))
Drawing_uncolored_circle = plt.Circle((0, 0 ), 1, fill=False )
# plt.scatter(*PointHandler.get_point(), color='b')
point_1 = PointHandler.get_point()
point_2 = PointHandler.get_point()
current_norm = norm(point_1, point_2)
current_color = color(current_norm)
plt.plot((point_1[0], point_2[0]), (point_1[1], point_2[1]), color=current_color)
plt.plot(TRIANGLE_POINTS["x"], TRIANGLE_POINTS["y"], color='r')
plt.xlim(-1.2,1.2)
plt.ylim(-1.2,1.2)
# axes.set_aspect( 1 )
axes.add_artist( Drawing_uncolored_circle )
plt.show()
```
